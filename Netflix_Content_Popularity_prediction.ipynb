# **Data Preprocessing**

1. Drop the unimportant cols[ Overview  ,  Original_Language , Poster-url]
2. Genre col has comma separeted values and white spaces that needs to be handled and casted catagory
3. Taking only year from the date


```python
#convet the date to datetime format

df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtype)
```

```python
#only taking the year from the data

df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.head()
```

```python
print(df.columns)
```

```python
cols = ['Overview', 'Original_Language', 'Poster_Url']
cols_to_drop = [col for col in cols if col in df.columns]
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
df.head()
```

**categorize Vote_Average column**
popular
average
below average
not popular

```python
def categorize_col(df, col, labels):
  edges =[df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]

  df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
  return df
```

```python
#defines level for edges
labels = ['not_popular', 'below_average', 'average', 'popular']
categorize_col(df, 'Vote_Average', labels)
df['Vote_Average']
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['Genre'] = df['Genre'].str.split(', ') #split the string into lists
df = df.explode('Genre')
df.head()
```

```python
df['Genre'] = df['Genre'].astype('category')#casting col in category
df['Genre'].dtypes
```
